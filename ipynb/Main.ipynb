Smote数据平衡
参数设置：
smote(x，y，sampling_strategy=“自动”，k_neighbors)


其中sampling_strategy：

- 'minority'：仅对少数类进行重新采样;
- 'not minority'：对除少数类以外的所有类进行重新采样;
- 'not majority'：重新采样除多数类之外的所有类;
- 'all'：对所有类进行重新采样;
- 'auto'：等效于 。'not majority'

```python
#ID_TC的SMOTE平衡
X1,y1=data_blance.smote(X_dataset_ID_TC, y_dataset_ID_TC,5,"not majority")

#ID_UC的SMOTE平衡
X2,y2=data_blance.smote(X_dataset_ID_UC, y_dataset_ID_UC,5,"not majority")

#TC_CC的SMOTE平衡
X3,y3=data_blance.smote(X_dataset_TC_CC, y_dataset_TC_CC,5,"not majority")

#UC_CC的SMOTE平衡
X4,y4=data_blance.smote(X_dataset_UC_CC, y_dataset_UC_CC,5,"not majority")

#UC_TC的SMOTE平衡
X5,y5=data_blance.smote(X_dataset_UC_TC, y_dataset_UC_TC,5,"not majority")

```
